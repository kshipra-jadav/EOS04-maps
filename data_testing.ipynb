```python
df = pd.read_csv("sentinel_sm.csv", )

e_indices = df[df['Sample Id (Grid)'].str.endswith('E', na=False)].index

rows_to_drop = []

for idx in e_indices:
    # Rows from idx to idx + 2
    sm_values = df.loc[idx:idx+2, 'SM1 (%)'].astype(float)
    mean_value = sm_values.mean()
    df.at[idx, 'SM1 (%)'] = mean_value
    rows_to_drop.extend([idx + 1, idx + 2])

# Drop the following two rows after each 'E'
df_cleaned = df.drop(rows_to_drop).reset_index(drop=True)

```

```python
df_cleaned
```

```python
import folium

center = df_cleaned[['Latitude (Centre of grid)', 'Longitude (Centre of grid)']].values[0]

m = folium.Map(location=center, zoom_start=15)

for lat, lon, label in df_cleaned[['Latitude (Centre of grid)', 'Longitude (Centre of grid)', 'Sample Id (Grid)']].values:
    folium.CircleMarker(
        location=(lat, lon),
        radius=5,
        color='blue',
        fill=True,
        fill_color='cyan',
        fill_opacity=0.7,
        popup=folium.Popup(str(label), parse_html=True)
    ).add_to(m)

m
```
